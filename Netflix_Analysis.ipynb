```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('netflix.csv')
df.head(20)
```

# Data Cleaning

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

# Type Wise Distribution

```python
type_indx=df['type'].value_counts().index
type_val=df['type'].value_counts().values
```

```python
plt.figure(figsize=(3,4))
plt.style.use('ggplot')
x=plt.bar(type_indx,type_val,width=0.2)
plt.bar_label(x,labels=type_val,label_type='edge')
plt.title('Type Wise Distribution')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

# Country Wise Distribution

```python
country_df=df['country'].value_counts().head(5)
```

```python
count = country_df.index
val = country_df.values
```

```python
plt.figure(figsize=(3,3))
plt.pie(val,labels=count,autopct='%1.1f%%',wedgeprops={'linewidth':0.5,'edgecolor':'black'})
plt.title('Country Wise Distribution(Top 5)')
plt.show()
```
